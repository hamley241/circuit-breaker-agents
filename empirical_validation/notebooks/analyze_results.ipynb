```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

summary_path = Path('../outputs/summary_REPLACE_ME.csv')
df = pd.read_csv(summary_path)
df
```

```python
plot_df = df.copy()
ax = plot_df.plot(x='cb_type', y='cascade_rate', kind='bar', legend=False, rot=0)
ax.set_ylabel('Cascade Rate')
ax.set_title('Cascade rate by breaker type')
plt.tight_layout()
plt.show()
```
